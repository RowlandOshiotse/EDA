```python
#From our initial boxplot for price, we could see that we have outliers in our data set
#Outliers are extreme values in our data set, they usually will alter our analysis
#So its best practice to remove outliers, there are number of ways to do this, i have decied to use the quantile method
#We start the process by definning our lower limit, then our upper limit and the find the interquartile range
Q1 = df.quantile(0.25)
```

```python
#Our lower limit, called Q1, or you can call it our 1st quartile
Q1
```

```python
#Our upper limit, called Q3, or you call it our 3rd quartile
Q3 =df.quantile(0.75)
```

```python
Q3
```

```python
#Interquartile range
#This subtracts our upper limit from our lower limit and gives us our interquartile or middle range
IQR = Q3 - Q1
```

```python
IQR
```

```python
#Now we remove the outliers using the below
#We simply saying here that any values less that our first quartile(lower limit) and any values higher than our third quartile,
#should be isolated
df = df[~((df< (Q1 -1.5 * IQR))   | (df> (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
#After removing the outlier, we can see the new shape our data
df.shape
```
